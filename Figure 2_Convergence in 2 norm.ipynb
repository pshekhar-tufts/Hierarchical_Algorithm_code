# Convergence in 2 norm in $R^n$

This notebook produces Figure 2 and analyzes the convergence behavior of the algorithm on the 4 test function in $R^n$

```python
from matplotlib.pyplot import * 
from Hierarchical_approach import figure2
import pickle
```

## 1.Gramacy and Lee

```python
Data = pickle.load( open( "gramacy.p", "rb" ) )
x1,y1,z1,s_conv1,s_crit1,coor1,epsilon1,conv_res1 = figure2(Data,err = 1.0e-2)
```

## 2. 1-D Schwefel function

```python
Data = pickle.load( open( "Schwefel.p", "rb" ) )
x2,y2,z2,s_conv2,s_crit2,coor2,epsilon2,conv_res2 = figure2(Data,err = 1.0e-2)
```

## 3. Dropwave function

```python
Data = pickle.load( open( "Dropwave.p", "rb" ) )
x3,y3,z3,s_conv3,s_crit3,coor3,epsilon3,conv_res3 = figure2(Data,err = 1.0e-2)
```

## 4. Schewel 2D

```python
Data = pickle.load( open( "Schwefel2d.p", "rb" ))
x4,y4,z4,s_conv4,s_crit4,coor4,epsilon4,conv_res4 = figure2(Data,err = 1.0e-2)
```
