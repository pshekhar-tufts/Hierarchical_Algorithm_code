# Reconstruction of bivariate function at different scales from $D^s_{sparse}$

```python
from matplotlib.pyplot import *
import pickle
from Hierarchical_approach import figure_4_5_7_9,Hierarchical_predict_basic
from mpl_toolkits.mplot3d import Axes3D
```

# 2-D Drop wave function

```python
Data = pickle.load( open( "Dropwave.p", "rb" ) )
fig = figure(figsize=(20,15))

s = 0
ax = fig.add_subplot(4, 2, 1)
x1,y1,z1,coor,epsilon,permv = figure_4_5_7_9(Data,s)
ax.scatter(x1,y1,color = 'r')
ax.scatter(Data[:,0],Data[:,1],color = 'b',s = 1)
####################################################
###### Predicting at new points
####################################################
m1 = 80
m2 = 80
xpred = np.linspace(-1,1,m1)
ypred = np.linspace(-1,1,m2)

Data_pred = []
for i in xpred:
    for j in ypred:
        Data_pred.append([i,j])


Data_pred = np.array(Data_pred)
f_star,basis = Hierarchical_predict_basic(x1,y1,coor,epsilon,Data_pred)
ax.set_title('(a) Scale 0: Sparse Representation, Rank: '+str(basis.shape[1]),size = 20)
tick_params(labelsize=15)
ax.set_xticks([-1,-0.5,0,0.5,1])
ax.set_yticks([-1,-0.5,0,0.5,1])
ax.set_xlabel('X',size = 20)
ax.set_ylabel('Y',size = 20)
####################################################
## Plotting the predicted fit
####################################################
Z3 = np.zeros([m1,m2])
counter = 0
X3 = sorted(set(Data_pred[:,0]))
Y3 = sorted(set(Data_pred[:,1]))
X3,Y3 = np.meshgrid(X3,Y3)
for i in range(m1):
    for j in range(m2):
        Z3[i,j] = f_star[counter]
        counter = counter + 1


ax = fig.add_subplot(4, 2, 2, projection='3d') 
# Plot the surface.
surf = ax.plot_surface(X3, Y3, Z3, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
#fig.colorbar(surf, shrink=0.5, aspect=10)
#ax.set_title('Test 3: Dropwave Function',size= 30)
ax.set_title('(b) Scale 0: Reconstruction from $D_{sparse}$ '+'('+str(basis.shape[1])+'/'+str(Data.shape[0])+')', size = 20)
tick_params(labelsize=10)
ax.set_xlabel('X',size = 20,labelpad=20)
ax.set_ylabel('Y',size = 20,labelpad=20)
ax.set_zlabel('Z',size = 20,labelpad=10)
ax.view_init(40, 20)
#ax.set_zlim([-1.1,0.5])
ax.set_xticks([-1,-0.5,0,0.5,1])
ax.set_yticks([-1,-0.5,0,0.5,1])
ax.set_zticks([-0.5,-0.3,-0.17])
tick_params(labelsize=15)







s = 2
ax = fig.add_subplot(4, 2, 3)
x1,y1,z1,coor,epsilon,permv = figure_4_5_7_9(Data,s)
ax.scatter(x1,y1,color = 'r')
ax.scatter(Data[:,0],Data[:,1],color = 'b',s = 1)

####################################################
###### Predicting at new points
####################################################
m1 = 80
m2 = 80
xpred = np.linspace(-1,1,m1)
ypred = np.linspace(-1,1,m2)

Data_pred = []
for i in xpred:
    for j in ypred:
        Data_pred.append([i,j])


Data_pred = np.array(Data_pred)
f_star,basis = Hierarchical_predict_basic(x1,y1,coor,epsilon,Data_pred)
ax.set_title('(c) Scale 2: Sparse Representation, Rank: '+str(basis.shape[1]),size = 20)
tick_params(labelsize=15)
ax.set_xticks([-1,-0.5,0,0.5,1])
ax.set_yticks([-1,-0.5,0,0.5,1])
ax.set_xlabel('X',size = 20)
ax.set_ylabel('Y',size = 20)
####################################################
## Plotting the predicted fit
####################################################
Z3 = np.zeros([m1,m2])
counter = 0
X3 = sorted(set(Data_pred[:,0]))
Y3 = sorted(set(Data_pred[:,1]))
X3,Y3 = np.meshgrid(X3,Y3)
for i in range(m1):
    for j in range(m2):
        Z3[i,j] = f_star[counter]
        counter = counter + 1


ax = fig.add_subplot(4, 2, 4, projection='3d') 
# Plot the surface.
surf = ax.plot_surface(X3, Y3, Z3, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
#fig.colorbar(surf, shrink=0.5, aspect=10)
#ax.set_title('Test 3: Dropwave Function',size= 30)
ax.set_title('(d) Scale 2: Reconstruction from $D_{sparse}$ '+'('+str(basis.shape[1])+'/'+str(Data.shape[0])+')', size = 20)
tick_params(labelsize=10)
ax.set_xlabel('X',size = 20,labelpad=20)
ax.set_ylabel('Y',size = 20,labelpad=20)
ax.set_zlabel('Z',size = 20,labelpad=10)
ax.view_init(40, 20)
#ax.set_zlim([-1.1,0.5])
ax.set_xticks([-1,-0.5,0,0.5,1])
ax.set_yticks([-1,-0.5,0,0.5,1])
ax.set_zticks([-0.55,-0.2,0])
tick_params(labelsize=15)





s = 4
ax = fig.add_subplot(4, 2, 5)
x1,y1,z1,coor,epsilon,permv = figure_4_5_7_9(Data,s)
ax.scatter(x1,y1,color = 'r')
ax.scatter(Data[:,0],Data[:,1],color = 'b',s = 1)

####################################################
###### Predicting at new points
####################################################
m1 = 80
m2 = 80
xpred = np.linspace(-1,1,m1)
ypred = np.linspace(-1,1,m2)

Data_pred = []
for i in xpred:
    for j in ypred:
        Data_pred.append([i,j])


Data_pred = np.array(Data_pred)
f_star,basis = Hierarchical_predict_basic(x1,y1,coor,epsilon,Data_pred)
ax.set_title('(e) Scale 4: Sparse Representation, Rank: '+str(basis.shape[1]),size = 20)
tick_params(labelsize=15)
ax.set_xticks([-1,-0.5,0,0.5,1])
ax.set_yticks([-1,-0.5,0,0.5,1])
ax.set_xlabel('X',size = 20)
ax.set_ylabel('Y',size = 20)
####################################################
## Plotting the predicted fit
####################################################
Z3 = np.zeros([m1,m2])
counter = 0
X3 = sorted(set(Data_pred[:,0]))
Y3 = sorted(set(Data_pred[:,1]))
X3,Y3 = np.meshgrid(X3,Y3)
for i in range(m1):
    for j in range(m2):
        Z3[i,j] = f_star[counter]
        counter = counter + 1


ax = fig.add_subplot(4, 2, 6, projection='3d') 
# Plot the surface.
surf = ax.plot_surface(X3, Y3, Z3, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
#fig.colorbar(surf, shrink=0.5, aspect=10)
#ax.set_title('Test 3: Dropwave Function',size= 30)
ax.set_title('(f) Scale 4: Reconstruction from $D_{sparse}$ '+'('+str(basis.shape[1])+'/'+str(Data.shape[0])+')', size = 20)
tick_params(labelsize=10)
ax.set_xlabel('X',size = 20,labelpad=20)
ax.set_ylabel('Y',size = 20,labelpad=20)
ax.set_zlabel('Z',size = 20,labelpad=10)
ax.view_init(40, 20)
#ax.set_zlim([-1.1,0.5])
ax.set_xticks([-1,-0.5,0,0.5,1])
ax.set_yticks([-1,-0.5,0,0.5,1])
ax.set_zticks([-0.86,-0.2,0.42])
tick_params(labelsize=15)







s = 6
ax = fig.add_subplot(4, 2, 7)
x1,y1,z1,coor,epsilon,permv = figure_4_5_7_9(Data,s)
ax.scatter(x1,y1,color = 'r')
ax.scatter(Data[:,0],Data[:,1],color = 'b',s = 1)

####################################################
###### Predicting at new points
####################################################
m1 = 80
m2 = 80
xpred = np.linspace(-1,1,m1)
ypred = np.linspace(-1,1,m2)

Data_pred = []
for i in xpred:
    for j in ypred:
        Data_pred.append([i,j])


Data_pred = np.array(Data_pred)
f_star,basis = Hierarchical_predict_basic(x1,y1,coor,epsilon,Data_pred)
ax.set_title('(g) Scale 6: Sparse Representation, Rank: '+str(basis.shape[1]),size = 20)
tick_params(labelsize=15)
ax.set_xticks([-1,-0.5,0,0.5,1])
ax.set_yticks([-1,-0.5,0,0.5,1])
ax.set_xlabel('X',size = 20)
ax.set_ylabel('Y',size = 20)
####################################################
## Plotting the predicted fit
####################################################
Z3 = np.zeros([m1,m2])
counter = 0
X3 = sorted(set(Data_pred[:,0]))
Y3 = sorted(set(Data_pred[:,1]))
X3,Y3 = np.meshgrid(X3,Y3)
for i in range(m1):
    for j in range(m2):
        Z3[i,j] = f_star[counter]
        counter = counter + 1


ax = fig.add_subplot(4, 2, 8, projection='3d') 
# Plot the surface.
surf = ax.plot_surface(X3, Y3, Z3, rstride=1, cstride=1,
                cmap='viridis', edgecolor='none')

ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
#fig.colorbar(surf, shrink=0.5, aspect=10)
#ax.set_title('Test 3: Dropwave Function',size= 30)
ax.set_title('(h) Scale 6: Reconstruction from $D_{sparse}$ '+'('+str(basis.shape[1])+'/'+str(Data.shape[0])+')', size = 20)
tick_params(labelsize=10)
ax.set_xlabel('X',size = 20,labelpad=20)
ax.set_ylabel('Y',size = 20,labelpad=20)
ax.set_zlabel('Z',size = 20,labelpad=10)
ax.view_init(40, 20)
#ax.set_zlim([-1.1,0.5])
ax.set_xticks([-1,-0.5,0,0.5,1])
ax.set_yticks([-1,-0.5,0,0.5,1])
ax.set_zticks([-1.02,-0.5,0])
tick_params(labelsize=15)

subplots_adjust(left=0.09, bottom = 0, right = 0.94, top = 0.99,wspace = 0.1,hspace = 0.5)   # determines the padding on the graph
show()
```
