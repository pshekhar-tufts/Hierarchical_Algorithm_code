# Reconstruction of univariate function at different scales from $D^s_{sparse}$

```python
from matplotlib.pyplot import *
import pickle
from Hierarchical_approach import figure_4_5_7_9,Hierarchical_predict_basic
```

# 1 D: Gramacy and Lee

```python
## Scale useful 0,2,4,6
Data = pickle.load( open( "Schwefel.p", "rb" ) )

figure(figsize=(20,12))


ax = subplot2grid((4,2),(0,0))
s = 0
x1,y1,z1,coor,epsilon,permv = figure_4_5_7_9(Data,s)
num = 200
Data_pred = np.zeros([num,2])
Data_pred[:,0] = np.linspace(Data[0,0],Data[-1,0],num).flatten()
f_star,basis = Hierarchical_predict_basic(x1,y1,coor,epsilon,Data_pred)
for u in range(basis.shape[1]):
    ax.plot(Data_pred[:,0],basis[:,u])
    
ax.set_title('(a) Scale 0: Spanning Basis, Rank: '+str(basis.shape[1]),size = 20)
tick_params(labelsize=15)
ax = subplot2grid((4,2),(0,1))
ax.scatter(Data[:,0],Data[:,2],marker = '*',s=1,color = 'g')
ax.scatter(x1,z1,s = 50)
ax.plot(Data_pred[:,0],f_star,color = 'r',lw=2)
ax.set_title('(b) Scale 0: Reconstruction from $D_{sparse}$ '+'('+str(basis.shape[1])+'/'+str(Data.shape[0])+')', size = 20)
ax.grid(True)
tick_params(labelsize=15)

ax = subplot2grid((4,2),(1,0))
s = 2
x1,y1,z1,coor,epsilon,permv = figure_4_5_7_9(Data,s)
num = 200
Data_pred = np.zeros([num,2])
Data_pred[:,0] = np.linspace(Data[0,0],Data[-1,0],num).flatten()
f_star,basis = Hierarchical_predict_basic(x1,y1,coor,epsilon,Data_pred)
for u in range(basis.shape[1]):
    ax.plot(Data_pred[:,0],basis[:,u])
ax.set_title('(c) Scale 2: Spanning Basis, Rank: '+str(basis.shape[1]),size = 20)
tick_params(labelsize=15)
ax = subplot2grid((4,2),(1,1))
ax.scatter(Data[:,0],Data[:,2],marker = '*',s=1,color = 'g')
ax.scatter(x1,z1,s = 50)
ax.plot(Data_pred[:,0],f_star,color = 'r',lw=2)
ax.set_title('(d) Scale 2: Reconstruction from $D_{sparse}$ '+'('+str(basis.shape[1])+'/'+str(Data.shape[0])+')', size = 20)
ax.grid(True)
tick_params(labelsize=15)


ax = subplot2grid((4,2),(2,0))
s = 4
x1,y1,z1,coor,epsilon,permv = figure_4_5_7_9(Data,s)
num = 200
Data_pred = np.zeros([num,2])
Data_pred[:,0] = np.linspace(Data[0,0],Data[-1,0],num).flatten()
f_star,basis = Hierarchical_predict_basic(x1,y1,coor,epsilon,Data_pred)
for u in range(basis.shape[1]):
    ax.plot(Data_pred[:,0],basis[:,u])
ax.set_title('(e) Scale 4: Spanning Basis, Rank: '+str(basis.shape[1]),size = 20)
tick_params(labelsize=15)
ax = subplot2grid((4,2),(2,1))
ax.scatter(Data[:,0],Data[:,2],marker = '*',s=1,color = 'g')
ax.scatter(x1,z1,s = 50)
ax.plot(Data_pred[:,0],f_star,color = 'r',lw=2)
ax.set_title('(f) Scale 4: Reconstruction from $D_{sparse}$ '+'('+str(basis.shape[1])+'/'+str(Data.shape[0])+')', size = 20)
ax.grid(True)
tick_params(labelsize=15)

ax = subplot2grid((4,2),(3,0))
s = 6
x1,y1,z1,coor,epsilon,permv = figure_4_5_7_9(Data,s)
num = 200
Data_pred = np.zeros([num,2])
Data_pred[:,0] = np.linspace(Data[0,0],Data[-1,0],num).flatten()
f_star,basis = Hierarchical_predict_basic(x1,y1,coor,epsilon,Data_pred)
for u in range(basis.shape[1]):
    ax.plot(Data_pred[:,0],basis[:,u])
ax.set_title('(g) Scale 6: Spanning Basis, Rank: '+str(basis.shape[1]),size = 20)
tick_params(labelsize=15)
ax = subplot2grid((4,2),(3,1))
ax.scatter(Data[:,0],Data[:,2],marker = '*',s=1,color = 'g')
ax.scatter(x1,z1,s = 50)
ax.plot(Data_pred[:,0],f_star,color = 'r',lw =2)
ax.set_title('(h) Scale 6: Reconstruction from $D_{sparse}$ '+'('+str(basis.shape[1])+'/'+str(Data.shape[0])+')', size = 20)
ax.grid(True)
tick_params(labelsize=15)
subplots_adjust(left=0.09, bottom = 0, right = 0.94, top = 0.99,wspace = 0.1,hspace = 0.3)   # determines the padding on the graph
show()
```
