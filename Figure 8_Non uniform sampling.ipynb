# Hierarchical Algorithm for learning on Non-uniformly sampled data

```python
from matplotlib.pyplot import *
from Hierarchical_approach import Hierarchical_learn_basic,Hierarchical_predict_basic
import pickle
```

Loading data from pickle file for just visualization of the true function. All the computations would be carried out on non-uniformly sampled data

```python
Data_gramacy = pickle.load( open( "gramacy.p", "rb" ) )
n = 200
x1 = np.linspace(0.5,2.5,n)
y1 = []
for i in x1:
    temp = (np.sin(10*np.pi*i))/(2*i) + (i-1)**4
    y1.append(temp)
flagx = max(max(x1),abs(min(x1)))
flagy = max(max(y1),abs(min(y1)))
```

## Showing 4 cases for non-uniform sampling of the data points

```python
## Gramacy and Lee

figure(figsize=(20,15))
##########################################################################
ax = subplot2grid((3,2),(0,0))
n = 40
x1 = sorted(0.51 + np.random.random(n-2)*1.98)
x1.insert(0, 0.5);x1.append(2.5)    ## So that we dont go into extrapolation
y1 = []
for i in x1:
    temp = (np.sin(10*np.pi*i))/(2*i) + (i-1)**4
    y1.append(temp)

Data = np.zeros([len(x1),3])
Data[:,0] = x1/flagx
Data[:,2] = y1/flagy
err = 10e-4
x1,y1,z1,s,coor,epsilon = Hierarchical_learn_basic(Data,err)
Data_pred = np.zeros([200,2])
Data_pred[:,0] = np.linspace(Data[0,0],Data[-1,0],200)
f_star,gauss_vec= Hierarchical_predict_basic(x1,y1,coor,epsilon,Data_pred)
ax.plot(Data_gramacy[:,0],Data_gramacy[:,2],color = 'k', label = 'True Function',lw = 3)
ax.plot(Data_pred[:,0],f_star,label = '$A_{S_a}f$',linestyle = '-.',color = 'g', lw = 4)
ax.scatter(Data[:,0],Data[:,2],label = 'Points not chosen',s = 100,color = 'b')
ax.scatter(x1,z1,color = 'r',label = '$D_{sparse}$',s = 100)
ax.set_title('(a): Reconstruction from $D_{sparse}$ '+'('+str(len(x1[0,:]))+'/'+str(Data.shape[0])+')',size = 25)
ax.grid(True)
ax.legend(fontsize=23)
tick_params(labelsize=20)

###########
ax = subplot2grid((3,2),(0,1))
n = 50
x1 = sorted(0.51 + np.random.random(n-2)*1.98)
x1.insert(0, 0.5);x1.append(2.5)    ## So that we dont go into extrapolation
y1 = []
for i in x1:
    temp = (np.sin(10*np.pi*i))/(2*i) + (i-1)**4
    y1.append(temp)

Data = np.zeros([len(x1),3])
Data[:,0] = x1/flagx
Data[:,2] = y1/flagy
err = 10e-4
x1,y1,z1,s,coor,epsilon = Hierarchical_learn_basic(Data,err)
Data_pred = np.zeros([200,2])
Data_pred[:,0] = np.linspace(Data[0,0],Data[-1,0],200)
f_star,gauss_vec= Hierarchical_predict_basic(x1,y1,coor,epsilon,Data_pred)
ax.plot(Data_gramacy[:,0],Data_gramacy[:,2],color = 'k', label = 'True Function',lw = 3)
ax.plot(Data_pred[:,0],f_star,label = '$A_{S_a}f$',linestyle = '-.',color = 'g', lw = 4)
ax.scatter(Data[:,0],Data[:,2],label = 'Points not chosen',s = 100,color = 'b')
ax.scatter(x1,z1,color = 'r',label = '$D_{sparse}$',s = 100)
ax.set_title('(b): Reconstruction from $D_{sparse}$ '+'('+str(len(x1[0,:]))+'/'+str(Data.shape[0])+')',size = 25)
ax.grid(True)
ax.legend(fontsize=23)
tick_params(labelsize=20)
##########################################################################
ax = subplot2grid((3,2),(1,0))
n = 60
x1 = sorted(0.51 + np.random.random(n-2)*1.98)
x1.insert(0, 0.5);x1.append(2.5)    ## So that we dont go into extrapolation
y1 = []
for i in x1:
    temp = (np.sin(10*np.pi*i))/(2*i) + (i-1)**4
    y1.append(temp)

Data = np.zeros([len(x1),3])
Data[:,0] = x1/flagx
Data[:,2] = y1/flagy
err = 10e-4
x1,y1,z1,s,coor,epsilon = Hierarchical_learn_basic(Data,err)
Data_pred = np.zeros([200,2])
Data_pred[:,0] = np.linspace(Data[0,0],Data[-1,0],200)
f_star,gauss_vec= Hierarchical_predict_basic(x1,y1,coor,epsilon,Data_pred)
ax.plot(Data_gramacy[:,0],Data_gramacy[:,2],color = 'k', label = 'True Function',lw = 3)
ax.plot(Data_pred[:,0],f_star,label = '$A_{S_a}f$',linestyle = '-.',color = 'g', lw = 4)
ax.scatter(Data[:,0],Data[:,2],label = 'Points not chosen',s = 100,color = 'b')
ax.scatter(x1,z1,color = 'r',label = '$D_{sparse}$',s = 100)
ax.set_title('(c): Reconstruction from $D_{sparse}$ '+'('+str(len(x1[0,:]))+'/'+str(Data.shape[0])+')',size = 25)
ax.grid(True)
ax.legend(fontsize=23)
tick_params(labelsize=20)
###########
ax = subplot2grid((3,2),(1,1))
n = 70
x1 = sorted(0.51 + np.random.random(n-2)*1.98)
x1.insert(0, 0.5);x1.append(2.5)    ## So that we dont go into extrapolation
y1 = []
for i in x1:
    temp = (np.sin(10*np.pi*i))/(2*i) + (i-1)**4
    y1.append(temp)

Data = np.zeros([len(x1),3])
Data[:,0] = x1/flagx
Data[:,2] = y1/flagy
err = 10e-4
x1,y1,z1,s,coor,epsilon = Hierarchical_learn_basic(Data,err)
Data_pred = np.zeros([200,2])
Data_pred[:,0] = np.linspace(Data[0,0],Data[-1,0],200)
f_star,gauss_vec= Hierarchical_predict_basic(x1,y1,coor,epsilon,Data_pred)
ax.plot(Data_gramacy[:,0],Data_gramacy[:,2],color = 'k', label = 'True Function',lw = 3)
ax.plot(Data_pred[:,0],f_star,label = '$A_{S_a}f$',linestyle = '-.',color = 'g', lw = 4)
ax.scatter(Data[:,0],Data[:,2],label = 'Points not chosen',s = 100,color = 'b')
ax.scatter(x1,z1,color = 'r',label = '$D_{sparse}$',s = 100)
ax.set_title('(d): Reconstruction from $D_{sparse}$ '+'('+str(len(x1[0,:]))+'/'+str(Data.shape[0])+')',size = 25)
ax.grid(True)
ax.legend(fontsize=23)
tick_params(labelsize=20)
######################################################################
subplots_adjust(left=0.09, bottom = 0, right = 0.94, top = 0.99,wspace = 0.25,hspace = 0.3)   # determines the padding on the graph


show()

```
