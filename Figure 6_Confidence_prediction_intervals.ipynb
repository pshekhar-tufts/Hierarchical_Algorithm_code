# Confidence and prediction bounds

```python
from matplotlib.pyplot import *
from Hierarchical_approach import figure6
import pickle
```

## 1-D Schewefel

```python
Data = pickle.load( open( "Schwefel.p", "rb" ) )
###########################
figure(figsize=(22,12))

ax = subplot2grid((2,3),(0,0))
s = 0
x = np.linspace(Data[0,0],Data[-1,0],300)
Data_pred = np.zeros([len(x),2])
Data_pred [:,0] = x.flatten()
mean_coor,std_coor_t,mean_y,std_y_t,std_pred_t,x1,y1,z1 = figure6(Data,Data_pred,s,confidence = 0.95)
ax.plot(x,mean_y,color = 'k',label = '$A_'+str(s)+'f$')
ax.fill_between(x.flatten(),mean_y.flatten()-std_y_t,mean_y.flatten()+std_y_t, alpha = 0.5,label = '95% t-confidence')
ax.fill_between(x.flatten(),mean_y.flatten()-std_pred_t,mean_y.flatten()+std_pred_t, alpha = 0.2,label = '95% t-Prediction')
ax.scatter(Data[:,0],Data[:,2],alpha = 0.2, label = 'Original Data')
ax.scatter(x1,z1,s = 80, label = '$D_{sparse}$')
ax.set_title('Confidence Interval')
ax.set_title('(a): Scale:'+str(s), size = 25)
ax.grid(True)
#ax.legend(loc = 8,  fontsize = 'x-large')
tick_params(labelsize=25)
###########################
ax = subplot2grid((2,3),(0,1))
s = 1
x = np.linspace(Data[0,0],Data[-1,0],300)
Data_pred = np.zeros([len(x),2])
Data_pred [:,0] = x.flatten()
mean_coor,std_coor_t,mean_y,std_y_t,std_pred_t,x1,y1,z1 = figure6(Data,Data_pred,s,confidence = 0.95)
ax.plot(x,mean_y,color = 'k',label = '$A_'+str(s)+'f$')
ax.fill_between(x.flatten(),mean_y.flatten()-std_y_t,mean_y.flatten()+std_y_t, alpha = 0.5,label = '95% t-confidence')
ax.fill_between(x.flatten(),mean_y.flatten()-std_pred_t,mean_y.flatten()+std_pred_t, alpha = 0.2,label = '95% t-Prediction')
ax.scatter(Data[:,0],Data[:,2],alpha = 0.2, label = 'Original Data')
ax.scatter(x1,z1,s = 80, label = '$D_{sparse}$')
ax.set_title('Confidence Interval')
ax.set_title('(b): Scale:'+str(s), size = 25)
#ax.legend( loc = 8, fontsize = 'x-large')
ax.grid(True)
tick_params(labelsize=25)

###########################
ax = subplot2grid((2,3),(0,2))
s = 2
x = np.linspace(Data[0,0],Data[-1,0],300)
Data_pred = np.zeros([len(x),2])
Data_pred [:,0] = x.flatten()
mean_coor,std_coor_t,mean_y,std_y_t,std_pred_t,x1,y1,z1 = figure6(Data,Data_pred,s,confidence = 0.95)
ax.plot(x,mean_y,color = 'k',label = '$A_'+str(s)+'f$')
ax.fill_between(x.flatten(),mean_y.flatten()-std_y_t,mean_y.flatten()+std_y_t, alpha = 0.5,label = '95% t-confidence')
ax.fill_between(x.flatten(),mean_y.flatten()-std_pred_t,mean_y.flatten()+std_pred_t, alpha = 0.2,label = '95% t-Prediction')
ax.scatter(Data[:,0],Data[:,2],alpha = 0.2, label = 'Original Data')
ax.scatter(x1,z1,s=80, label = '$D_{sparse}$')
ax.set_title('Confidence Interval')
ax.set_title('(c): Scale:'+str(s), size = 25)
#ax.legend(loc = 8,  fontsize = 'x-large')
ax.grid(True)
tick_params(labelsize=25)

###########################
ax = subplot2grid((2,3),(1,0))
s = 3
x = np.linspace(Data[0,0],Data[-1,0],300)
Data_pred = np.zeros([len(x),2])
Data_pred [:,0] = x.flatten()
mean_coor,std_coor_t,mean_y,std_y_t,std_pred_t,x1,y1,z1 = figure6(Data,Data_pred,s,confidence = 0.95)
ax.plot(x,mean_y,color = 'k',label = '$A_'+str(s)+'f$')
ax.fill_between(x.flatten(),mean_y.flatten()-std_y_t,mean_y.flatten()+std_y_t, alpha = 0.5,label = '95% t-confidence')
ax.fill_between(x.flatten(),mean_y.flatten()-std_pred_t,mean_y.flatten()+std_pred_t, alpha = 0.2,label = '95% t-Prediction')
ax.scatter(Data[:,0],Data[:,2],alpha = 0.2, label = 'Original Data')
ax.scatter(x1,z1,s=80, label = '$D_{sparse}$')
ax.set_title('Confidence Interval')
ax.set_title('(d): Scale:'+str(s), size = 25)
#ax.legend(loc = 8,  fontsize = 'x-large')
ax.grid(True)
tick_params(labelsize=25)

###########################
ax = subplot2grid((2,3),(1,1))
s = 4
x = np.linspace(Data[0,0],Data[-1,0],300)
Data_pred = np.zeros([len(x),2])
Data_pred [:,0] = x.flatten()
mean_coor,std_coor_t,mean_y,std_y_t,std_pred_t,x1,y1,z1 = figure6(Data,Data_pred,s,confidence = 0.95)
ax.plot(x,mean_y,color ='k',label = '$A_sf$')
ax.fill_between(x.flatten(),mean_y.flatten()-std_y_t,mean_y.flatten()+std_y_t, alpha = 0.5,label = '95% t-confidence')
ax.fill_between(x.flatten(),mean_y.flatten()-std_pred_t,mean_y.flatten()+std_pred_t, alpha = 0.2,label = '95% t-Prediction')
ax.scatter(Data[:,0],Data[:,2],alpha = 0.2, label = 'Original Data')
ax.scatter(x1,z1,s=80, label = '$D^s_{sparse}$')
ax.set_title('Confidence Interval')
ax.set_title('(e): Scale:'+str(s), size = 25)
#ax.legend(loc = 8,  fontsize = 'x-large')
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1),
          fancybox=True, shadow=True, ncol=5,fontsize=26)
ax.grid(True)
tick_params(labelsize=25)


###########################
ax = subplot2grid((2,3),(1,2))
s = 5
x = np.linspace(Data[0,0],Data[-1,0],300)
Data_pred = np.zeros([len(x),2])
Data_pred [:,0] = x.flatten()
mean_coor,std_coor_t,mean_y,std_y_t,std_pred_t,x1,y1,z1 = figure6(Data,Data_pred,s,confidence = 0.95)
ax.plot(x,mean_y,color = 'k',label = '$A_'+str(s)+'f$')
ax.fill_between(x.flatten(),mean_y.flatten()-std_y_t,mean_y.flatten()+std_y_t, alpha = 0.5,label = '95% t-confidence')
ax.fill_between(x.flatten(),mean_y.flatten()-std_pred_t,mean_y.flatten()+std_pred_t, alpha = 0.2,label = '95% t-Prediction')
ax.scatter(Data[:,0],Data[:,2],alpha = 0.2, label = 'Original Data')
ax.scatter(x1,z1,s=80, label = '$D_{sparse}$')
ax.set_title('Confidence Interval')
tick_params(labelsize=25)
ax.set_title('(f): Scale:'+str(s), size = 25)
#ax.legend(loc = 8, fontsize = 'x-large')
ax.grid(True)
subplots_adjust(left=0.09, bottom = 0, right = 0.94, top = 0.99,wspace = 0.2,hspace = 0.2)   # determines the padding on the graph


show()


```
