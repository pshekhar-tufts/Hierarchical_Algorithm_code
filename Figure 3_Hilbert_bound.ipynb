# Plotting the upper bound of the inner prooduct in $\mathcal{H_s}$

```python
from matplotlib.pyplot import *
from Hierarchical_approach import figure3
import pickle
```

Importing the data from the pickle files 

```python
Data = pickle.load( open( "gramacy.p", "rb" ) )
bounds1 = figure3(Data)

Data = pickle.load( open( "Schwefel.p", "rb" ) )
bounds2 = figure3(Data)

Data = pickle.load( open( "Dropwave.p", "rb" ) )
bounds3 = figure3(Data)

Data = pickle.load( open( "Schwefel2d.p", "rb" ) )
bounds4 = figure3(Data)
```

Plotting the results using the bounds computed in function figure3

```python
figure(figsize=(20,10))
ax = subplot2grid((2,2),(0,0))
ax.set_yscale('log')
ax.plot(bounds1[:,0],bounds1[:,1],lw = 3,color = 'g')
ax.scatter(bounds1[:,0],bounds1[:,1],color = 'r',s = 300,marker = '*')
ax.grid(True)
ax.set_xlabel('Scale',size = 25)
ax.set_ylabel('$||C_s||_{\infty}||E^s|_X||_1$',size = 25)
ax.set_title('(a)',size = 28)
ax.fill_between(bounds1[:,0],bounds1[:,1],np.zeros([bounds1.shape[0],1]).flatten(),alpha = 0.4,color = 'y')
tick_params(labelsize=25)



ax = subplot2grid((2,2),(0,1))
ax.set_yscale('log')
ax.plot(bounds2[:,0],bounds2[:,1],lw = 3,color = 'g')
ax.scatter(bounds2[:,0],bounds2[:,1],color = 'r',s = 300,marker = '*')
ax.grid(True)
ax.set_xlabel('Scale',size = 25)
ax.set_ylabel('$||C_s||_{\infty}||E^s|_X||_1$',size = 25)
ax.set_title('(b)',size = 28)
ax.fill_between(bounds2[:,0],bounds2[:,1],np.zeros([bounds2.shape[0],1]).flatten(),alpha = 0.4,color = 'y')
tick_params(labelsize=25)




ax = subplot2grid((2,2),(1,0))
ax.set_yscale('log')
ax.plot(bounds3[:,0],bounds3[:,1],lw = 3,color = 'g')
ax.scatter(bounds3[:,0],bounds3[:,1],color = 'r',s = 300,marker = '*')
ax.grid(True)
ax.set_xlabel('Scale',size = 25)
ax.set_ylabel('$||C_s||_{\infty}||E^s|_X||_1$',size = 25)
ax.set_title('(c)',size = 28)
ax.fill_between(bounds3[:,0],bounds3[:,1],np.zeros([bounds3.shape[0],1]).flatten(),alpha = 0.4,color = 'y')
tick_params(labelsize=25)


ax = subplot2grid((2,2),(1,1))
ax.set_yscale('log')
ax.plot(bounds4[:,0],bounds4[:,1],lw = 3,color = 'g')
ax.scatter(bounds4[:,0],bounds4[:,1],color = 'r',s = 300,marker = '*')
ax.grid(True)
ax.set_xlabel('Scale',size = 25)
ax.set_ylabel('$||C_s||_{\infty}||E^s|_X||_1$',size = 25)
ax.set_title('(d)',size = 28)
ax.fill_between(bounds4[:,0],bounds4[:,1],np.zeros([bounds4.shape[0],1]).flatten(),alpha = 0.4,color = 'y')
tick_params(labelsize=25)
subplots_adjust(left=0.09, bottom = 0, right = 0.94, top = 0.99,wspace = 0.25,hspace = 0.3)   # determines the padding on the graph


show()
```
